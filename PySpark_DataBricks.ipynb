```python
from pyspark.ml.feature import VectorAssembler
feature_assmble=VectorAssembler(inputCols=['longitude',
 'latitude',
 'housing_median_age',
 'total_rooms',
 'total_bedrooms',
 'population',
 'households',
 'median_income'],outputCol="Independent_Features")
```

```python
output=feature_assmble.transform(df_spark)
output.show()
```

```python
final_Data=output.select("median_house_value","Independent_Features")
```

```python
final_Data.show()
```

```python

from pyspark.ml.regression import LinearRegression
train_data,test_data=final_Data.randomSplit([0.75,0.25])
```

```python
regressor=LinearRegression(featuresCol="Independent_Features",labelCol="median_house_value")
```

```python
regressor=regressor.fit(train_data)
```

```python
regressor.coefficients 
```

```python
regressor.intercept
```

```python
pred=regressor.evaluate(test_data)
```

```python
pred.predictions.show()
```

```python
pred.meanAbsoluteError,pred.meanSquaredError 
```
